```python
%matplotlib widget
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib.patches as patches
import numpy as np
import scipy
from scipy.optimize import curve_fit
import cv2
from PIL import Image
import glob

import images_library
```

```python
outputFile="buio2_11.txt"

xROI, yROI, rROI, signal, mean, stdev = np.loadtxt(outputFile, unpack=True)
time = np.linspace(0, len(xROI)/3-1, int(len(xROI)/3))
time
```

```python
maskNoROI= xROI== 0
mask100ROI= xROI== 100
mask400ROI= xROI== 400

fig, ax = plt.subplots(1,1, figsize=(3,3))
ax.errorbar(time, mean[maskNoROI], yerr=stdev[maskNoROI], fmt='.', label='all')
ax.errorbar(time, mean[mask100ROI], yerr=stdev[mask100ROI], fmt='.', label='center (100,100)')
ax.errorbar(time, mean[mask400ROI], yerr=stdev[mask400ROI], fmt='.', label='center (400, 400)')
ax.set_xlabel("time [n of acquisition]")
ax.set_ylabel("mean signal")
fig.suptitle("02/11/22") 
plt.legend()
```
