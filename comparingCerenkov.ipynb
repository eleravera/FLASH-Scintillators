```python
PMMAcentered = "/home/eleonora/FLASH-Scintillators/notes/txt/PMMAcentered.txt"
PMMAmoved = "/home/eleonora/FLASH-Scintillators/notes/txt/PMMAmoved.txt"
WithoutScintillator = "/home/eleonora/FLASH-Scintillators/notes/txt/WithoutScintillator.txt"
WithoutScintWithoutPMMA = "/home/eleonora/FLASH-Scintillators/notes/txt/WithoutScintillator-WithoutPMMA.txt"

PMMAcenteredData = np.transpose(np.loadtxt(PMMAcentered))
PMMAmovedData = np.transpose(np.loadtxt(PMMAmoved))
WithoutScintillatorData = np.transpose(np.loadtxt(WithoutScintillator))
WithoutScintWithoutPMMAData = np.transpose(np.loadtxt(WithoutScintWithoutPMMA))

radii= PMMAcenteredData[0]
```

```python
def ratioError(a, aErr, b, bErr):
    a = np.array(a)
    b=np.array(b)
    aErr = np.array(aErr)
    bErr=np.array(bErr)  
    error =np.sqrt((aErr/b)**2 + (bErr/b * a/b)**2) 
    return error

def ratio(a,b):
    a=np.array(a)
    b=np.array(b)
    return a/b


def plot(x, y, yNorm, yErr, yNormErr, label='', fmt='.', ylabel=''):
    fig, ax = plt.subplots(1,1, figsize=(7,5))
    y_ = ratio(y, yNorm)
    yerr_=ratioError(y, yErr, yNorm, yNormErr)
    ax.errorbar(x, y_, yerr_, fmt=fmt, label=label)
    ax.set_ylabel(ylabel, fontsize=16)
    ax.set_xlabel("Raggio ROI [n of pixel]", fontsize=16)
    #fig.legend(loc="upper center", fontsize=20)
  
```

```python
plot(radii, PMMAmovedData[1], PMMAcenteredData[1], PMMAmovedData[2], PMMAcenteredData[2],
     label='', fmt='.', ylabel='PMMA moved/ centered')

plot(radii, WithoutScintillatorData[1], PMMAcenteredData[1], WithoutScintillatorData[2], PMMAcenteredData[2],
     label='', fmt='.', ylabel='no scint/ centered')

plot(radii, WithoutScintWithoutPMMAData[1], PMMAcenteredData[1], WithoutScintWithoutPMMAData[2], PMMAcenteredData[2],
     label='', fmt='.', ylabel='no scint no PMMA/ centered')

```
